### Data Description

**ID:** Player's unique identifier.

**Name:** Player's abbreviated name.

**FullName:** Player's full name.

**Age:** Player's age.

**Height:** Player's height.

**Weight:** Player's weight.

**Nationality:** Player's nationality.

**Overall:** Player's overall statistics regarding their abilities.

**Potential:** Player's potential. 

**Growth:** Player's growth.

**BestPosition:** Position in the field that a player generally plays.

**Club:** The Club that the player plays for. 

**ValueEUR:** Value in Euro of the player.

**WageEUR:** Wage of the player in Euro.

Note this dataset was last updated in January, so information may be outdated or incorrect. 
Use the following two cells to make sure you are loading the dataset correctly.

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt

#reading dataset
df = pd.read_csv('players_fifa23.csv')
df
```

```python
#describing dataset
df.describe()
```

## Removing and Creating columns For analysis

```python
#Removing all columns
df = df.loc[:, :'WageEUR']
df.columns
```

```python
#Removing TotalStats, BaseStats, Positions and PhotoUrl
df = df.drop(['TotalStats','BaseStats','Positions','PhotoUrl'], axis = 1)
df.columns
```

```python
#Adding column GMValue that computes the following equation: $ 'Overall' * 0.9 + 'Growth'$
df['GMValue'] = pd.to_numeric(df['Overall'] * 0.9 + df['Growth'])
df.head()
```

```python
# Adding FirstName & LastName Columns
df[['FirstName','LastName']] = df['FullName'].str.split(' ', 1, expand=True)
df.head()
```

```python
df.GMValue.min() , df.GMValue.max()
```

```python
# Sorting Dataframe using GMValue in descending order
df = df.sort_values(by=['GMValue'], ascending=False)
df
```

3. What is the average potential and average overall score for each group age younger than 19 (including 19 years old players)? Could you find any trend? Please discuss

```python
#1. What are the 5 nations with the highest average overall score and that at least 50 players are listed in the dataset?

#creating a copy 
nation_with_high_avg_overall_with_fifty_player = df.copy()
#adding column in the newdataset
nation_with_high_avg_overall_with_fifty_player['Count'] = nation_with_high_avg_overall_with_fifty_player.groupby(['Nationality'])['Nationality'].transform('count')
#Finding the average overall 
nation_with_high_avg_overall_with_fifty_player = nation_with_high_avg_overall_with_fifty_player.groupby(['Nationality','Count'])['Overall'].mean().reset_index(name='AvgOverall')
#filtering nationality where minimum count is not 50
nation_with_high_avg_overall_with_fifty_player = nation_with_high_avg_overall_with_fifty_player[nation_with_high_avg_overall_with_fifty_player['Count']>=50]
#printing 5 nations with the highest average overall score and that at least 50 players are listed in the dataset
nation_with_high_avg_overall_with_fifty_player.sort_values(by = 'AvgOverall', ascending = False).head(5)
```

```python
# 2.Which 5 clubs are paying the highest total wage to their players?

#creating a copy
club_with_highest_salary = df.copy()
#creating a new column for salary by clubs
club_with_highest_salary = club_with_highest_salary.groupby(['Club'])['WageEUR'].sum().reset_index(name = 'TotalSalary')
club_with_highest_salary.sort_values(by = 'TotalSalary', ascending = False).head(5)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 10))
axs=axs.flatten() 

countries = ['England','Germany','Spain', 'Argentina']
bp_ds = df[df.Nationality.isin(countries)]

ds = bp_ds.assign(Count=1).groupby(['Nationality'])['Count'].count().reset_index()

bp_ds.boxplot(ax = axs[0], column=['Overall'], by = 'Nationality', rot = 90)
axs[0].set_title('Countrywise Players Overall Score Distribution')
axs[0].set_xlabel('Country Names')
axs[0].set_ylabel('Overall Score')
print('Before Removing : ')
print(bp_ds.groupby(['Nationality']).agg({'Overall' : ['mean']}))



overall_delete = bp_ds.Overall.between(bp_ds.Overall.quantile(.05), bp_ds.Overall.quantile(.95))
bp_ds = bp_ds[overall_delete]
bp_ds.boxplot(ax = axs[1], column=['Overall'], by = 'Nationality', rot=90)
axs[1].set_title('Players New Overall Score Distribution')
axs[1].set_xlabel('Country Name')
axs[1].set_ylabel('Overall Score')
print('\n\nAfter Removing')
print(bp_ds.groupby(['Nationality']).agg({'Overall' : ['mean']}))

plt.show()
```

### Plot 2

In this task, we are interested if certain countries are better at producing different kinds of players. For this exercise, we are not interested in the goal keeper. The X axis of your charts should be the Nationality of the player, and the bars should be split by position for each nationality. You can assume that players only play the position defined under BestPosition. You can ignore additional positions that are not defined below.

1. The first plot must be a grouped bar chart for Forwards for each Country. Positions (LW, ST, RW).

2. The second is a grouped bar chart for Midfielders. Positions (LM, RM, CM)

3. The third plot is a grouped bar chart for Defenders. Positions (LB, RB, CB)

4. The fourth plot is a grouped bar chart except instead of grouping by position, we will group by forwards, midfielders, and defenders. 

```python
import numpy as np

#creating subplots
fig, axs = plt.subplots(2, 2, figsize=(20, 10))
axs=axs.flatten() 

#defining condition based on dataset will be filtered and created
countries = ['England','Germany','Spain', 'Argentina']
ds_plt = df[df.Nationality.isin(countries)]

#conditions for position wise
forward = ['LW','ST','RW']
midfield = ['LM','RM','CM']
defender = ['LB','RB','CB']

#creating dataset based on position condition
forward_ds = ds_plt[ds_plt.BestPosition.isin(forward)]
midfield_ds = ds_plt[ds_plt.BestPosition.isin(midfield)]
defender_ds = ds_plt[ds_plt.BestPosition.isin(defender)]

#creating a new column Count in the positional dataset
forward_group = forward_ds.assign(Count=1).groupby(['Nationality','BestPosition'])['Count'].count().reset_index()
midfield_group = midfield_ds.assign(Count=1).groupby(['Nationality','BestPosition'])['Count'].count().reset_index()
defender_group = defender_ds.assign(Count=1).groupby(['Nationality','BestPosition'])['Count'].count().reset_index()


#creating bar charts with pivot table from the positional dataset
#1. Forward Barchart
forward_group.pivot_table(index='Nationality', 
                          columns='BestPosition', 
                          values='Count').plot(kind = 'bar',
                                              ax = axs[0],
                                              rot=0,
                                              title='Country Wise Forward Players Bar-chart')

#2. Midfield Barchart
midfield_group.pivot_table(index = 'Nationality',
                          columns = 'BestPosition',
                          values = 'Count').plot(kind = 'bar',
                                                ax = axs[1],
                                                rot = 0,
                                                title = 'Country Wise Midfielders Players Bar-chart')

#3. Defender Barchart
defender_group.pivot_table(index = 'Nationality',
                          columns = 'BestPosition',
                          values = 'Count').plot(kind = 'bar',
                                                ax = axs[2],
                                                rot = 0,
                                                title = 'Country Wise Defender Players Bar-chart')

#defining condition for positional group
conditions = [
    (ds_plt.BestPosition.isin(forward)),
    (ds_plt.BestPosition.isin(midfield)),
    (ds_plt.BestPosition.isin(defender))
]

#defining values for conditions
values = ['Forward','Midfielder','Defender']

#creating new column based on condition and setting value for the columns when condition are matched
ds_plt['PlayerGroup'] = np.select(conditions, values)

#filtering rows where our desired values are present
playergroup = ds_plt[ds_plt.PlayerGroup.isin(values)]
#Creating new column count 
playergroup = playergroup.assign(Count=1).groupby(['Nationality','PlayerGroup'])['Count'].count().reset_index()
#creating pivot table and plotting barchart with the pivot table
playergroup.pivot_table(index = 'Nationality',
                       columns = 'PlayerGroup',
                       values = 'Count').plot(kind = 'bar',
                                             ax = axs[3],
                                             rot = 0,
                                             title='Country Wise Player Group Bar-chart')




plt.show()
```
